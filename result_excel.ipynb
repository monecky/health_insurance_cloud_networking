```python
from collections import Counter

# Generate random data for the histogram
x = [df["Login cloud"],df["Registration cloud"] ,df["Home cloud"]]


# plt.hist(x, fill=True, histtype="bar", linewidth=10,color = ["orange", "tan", "red"],
#         label=["Login score", "Registration score","Home score"], bins=["Microsoft", "Akamai", "Amazon","Cloudflare","MWAY", "ASR","KPN", "Orange", "interconnect", "Claranet", "Google"])
# The labels for the x-axis
bins = ["Microsoft", "Akamai", "Cloudflare","Amazon",  "MWAY", "ASR", "KPN", "Orange", "interconnect", "Claranet", "Google"]

# Create the x positions for the bars
x = np.arange(len(bins))  # Create positions for each label
width = 0.25  # Set the width of the bars

# First set of bars: Login cloud
counter = Counter(df["Login cloud"])
counts = [counter.get(label, 0) for label in bins]
plt.bar(x - width, counts, width=width, color="orange", label="Login cloud")

# Second set of bars: Registration cloud
counter = Counter(df["Registration cloud"])
counts = [counter.get(label, 0) for label in bins]
plt.bar(x, counts, width=width, color="tan", label="Registration cloud")

# Third set of bars: Home cloud
counter = Counter(df["Home cloud"])
counts = [counter.get(label, 0) for label in bins]
plt.bar(x + width, counts, width=width, color="red", label="Home cloud")

# Adding labels and title
plt.xticks(x, bins, rotation=45)  # Set the x-axis labels
plt.xlabel('Provider')
plt.ylabel('Count')
plt.title('Who manages ISP and Cloud, from the company')

# Adding the legend
plt.legend()

# Display the plot
plt.show()
```
