```python
import pandas as pd

file_path = "data_health_insurance.xlsx"
df = pd.read_excel(file_path, sheet_name=0)
print(df.columns)
companies = {}
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
x = [df["Login score"],df["Registration score"] ,df["Home score"]]


plt.hist(x, 12, fill=True, histtype="bar", linewidth=5,color = ["orange", "tan", "red"],
        label=["Login score", "Registration score","Home score"])
# Adding labels and title
plt.xlabel('Score (in %)')
plt.ylabel('Count')
plt.title('Histogram internet.nl secure score')
plt.legend(["Login score", "Registration score","Home score"])
# plt.bar_label(["Login score", "Registration score","Home score"])
# Display the plot
plt.show()
```

```python
# Generate random data for the histogram
x = [df["Login cloud"],df["Registration cloud"] ,df["Home cloud"]]


plt.hist(x, 12, fill=True, histtype="bar", linewidth=10,color = ["orange", "tan", "red"],
        label=["Login score", "Registration score","Home score"])
# Adding labels and title
plt.xticks(rotation=45)
plt.xlabel('Provider')
plt.ylabel('Count')
plt.title('Who manage ISP and cloud, from the company')
plt.legend(["Login cloud", "Registration cloud","Home cloud"])
# plt.bar_label(["Login score", "Registration score","Home score"])
# Display the plot
plt.show()
```
